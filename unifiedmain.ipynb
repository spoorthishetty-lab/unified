```python
orders=df["Vehicle_Class"].value_counts()
print(orders)
x=orders.index
y=orders.values
print(x)
print(y)
plt.figure(figsize=(10,10),dpi=100)
plt.title("Orders based on Vehicle Class",fontdict={"family":"georgia","color":"red","weight":"bold","size":"20"})
plt.xlabel("Orders")
plt.ylabel("Vehicle_Class")
plt.hist(y)
plt.show()
```

```python
popular=df.groupby('Vehicle_Class')['Year'].sum().sort_values(ascending=False).head(5)
popular
popular.plot(kind='barh')
plt.title("Top Five Vehicle Class based on Year",fontdict={"family":"georgia","weight":"bold","size":20,"color":"Red"})
plt.show()
```

```python
popular=df.groupby('Vehicle_Class')['Year'].sum().sort_values(ascending=False).head(5)
popular
```

```python
orders=df['Vehicle_Category'].value_counts()
print(orders)
x=orders.values
print(x)
lb=orders.index
print(lb)
plt.title("Sales based on Vehicle category",fontdict={"family":"georgia","weight":"bold","size":"20","color":"red"})
e=[0.1,0,0,0,0]
c=["purple","yellow","blue","green","pink"]
plt.pie(x,labels=lb,autopct='%1.f%%',explode=e,colors=c)
plt.legend(lb,loc='upper right')
plt.show()
```

```python
orders=df['Month_Name'].value_counts()
print(orders)
x=orders.index
y=orders.values
print(x)
print(y)
plt.figure(figsize=(10,5),dpi=100)
plt.title("Sales based on Month",fontdict={"family":"georgia","color":"red","weight":"bold","size":20})
plt.xlabel("Month")
plt.ylabel("Orders")
c=["red","green","yellow","orange","purple","pink","blue","green","red","yellow","pink","blue"]
plt.bar(x,y,color=c)
plt.show()
```

```python
y=df.loc[300:350,'EV_Sales_Quantity']
#print(y)
plt.plot(y,'gs--')
plt.title("Distribution of Quantity",fontdict={"family":"georgia","color":"red","weight":"bold","size":20})
plt.xlabel("counts",color='RED')
plt.ylabel("Quantity",color='red')
plt.show()
```
